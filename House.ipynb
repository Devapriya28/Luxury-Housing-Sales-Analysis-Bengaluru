# CLEANING DATA

```python
# Convert cr to float

def convert_ticket_price(value):
    if isinstance(value, str):
        value = (
            value.replace("₹", "")   # remove rupee symbol
                 .replace("Cr", "")  # remove cr
                 .strip()
        )
        return float(value)
    return value

house_datafile["Ticket_Price_Cr"] = house_datafile["Ticket_Price_Cr"].apply(convert_ticket_price)

# Check cleaned values
house_datafile["Ticket_Price_Cr"].head()

```

```python
house_datafile["Ticket_Price_Cr"].sample(20)
```

```python
#fill missing values

house_datafile['Unit_Size_Sqft'] = house_datafile['Unit_Size_Sqft'].fillna(house_datafile['Unit_Size_Sqft'].mean())
house_datafile['Ticket_Price_Cr'] = house_datafile['Ticket_Price_Cr'].fillna(house_datafile['Ticket_Price_Cr'].mean())
house_datafile['Amenity_Score'] = house_datafile['Amenity_Score'].fillna(house_datafile['Amenity_Score'].median())
house_datafile['Buyer_Comments'] = house_datafile['Buyer_Comments'].fillna("No Comments")
```

```python
house_datafile["Price_per_Sqft"] = house_datafile["Ticket_Price_Cr"] * 10000000 / house_datafile["Unit_Size_Sqft"]
house_datafile.sample(10)
```

```python
house_datafile['Purchase_Quarter'] = pd.to_datetime(house_datafile['Purchase_Quarter'])
house_datafile['Quarter_Number'] = house_datafile['Purchase_Quarter'].dt.quarter
house_datafile['Quarter_Number']
```

```python
text_cols = ['Micro_Market', 'Developer_Name', 'Configuration',
             'Transaction_Type', 'Buyer_Type', 'Possession_Status',
             'Sales_Channel', 'NRI_Buyer']
for col in text_cols:
    house_datafile[col] = house_datafile[col].str.strip().str.title()
house_datafile
```

```python
house_datafile['Transaction_Type'].unique()
```

```python
#Primary = booked (1), Secondary = not booked (0)
house_datafile['Booking_Flag'] = house_datafile['Transaction_Type'].str.lower().map({'primary': 1, 'secondary': 0})
```
